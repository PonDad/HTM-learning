# HTMで音楽をおぼえる

Scratchsサイトにリンクして音源を再生します。

[サウンド・リピーター - Scratch](https://scratch.mit.edu/projects/390512046)

Chrome拡張機能を使って音源を録音します。

[AudioRecorder](https://chrome.google.com/webstore/detail/audiorecorder/enhfkjkjfhhdibpgjmiamdcdgmcjpplk?hl=ja)

HTMのサンプルアプリ`nupic.critic`をレポジトリからクローンして下さい。

[htm-community/nupic.critic](https://github.com/htm-community/nupic.critic)


以下のソフトが必要になりますので、Python2系のものをインストールして下さい。

- [NuPIC](https://github.com/numenta/nupic)
- [matplotlib](https://matplotlib.org/)
- [numpy](https://numpy.org/)

```python
!git clone https://github.com/htm-community/nupic.critic.git
```

先程ダウンロードしたwavファイルをカレントディレクトリに移動して下さい。

```python
!python nupic.critic/convert_wav.py myRecording01.wav
```

ディレクトリ内に`data/myRecording01_5hz_10b/input`フォルダが生成されました。この中にWAVファイルを音階ごとにデータ化したファイルが保存されています。

保存されたファイルをSDRに変換後、学習と予測を同時におこないます。

```python
!python nupic.critic/run.py data/myRecording01_5hz_10b/input
```

今度はディレクトリ内に`data/myRecording01_5hz_10b/input`フォルダが生成されました。この中に時間ごとの予測と、誤差が保存されています。

保存したデータをプロットして可視化出来ます。音楽の再生と合わせてどの部分が記憶できたか確認してみましょう。

音楽を同時再生する際、Macの`open`コマンドを使います。

Ubuntuで実行する際は`nupic.critic/plotter.py`98行目

`subprocess.call("open %s" % audio_file, shell=True)`を

`subprocess.call("xdg-open %s" % audio_file, shell=True)`に修正してください。

```python
!python nupic.critic/plotter.py data/myRecording01_5hz_10b/output -w 'myRecording01.wav'
```
